```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('beer_reviews.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.info()
```

```python
#order reviews and drop duplicates
df.sort_values('review_overall', ascending=False)
df = df.drop_duplicates(subset=['review_profilename', 'beer_name'], keep='first')
df.info()
```

```python
#Drop reviews of less than 1
df = df.loc[df.review_overall >= 1]
df.info()
```

```python
df.head()
```

```python
df.to_csv('beer_reviews_clean.csv')
```
